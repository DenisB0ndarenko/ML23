```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine = load_wine()
```

```python
wine_x_ds = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])
wine_x_ds
```

```python
wine_y_ds = pd.DataFrame(data=wine['target'])
wine_y_ds
```

```python
wine_ds = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])
wine_ds['target'] = wine['target']
wine_ds
```

```python
wine_ds.dtypes
```

```python
# Проверим наличие пустых значений
# Цикл по колонкам датасета
for col in wine_ds.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = wine_ds[wine_ds[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
wine_ds.duplicated().sum()
```

```python
sns.lmplot(x='flavanoids', y='total_phenols', data=wine_ds, fit_reg=False, hue='target', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='upper right', fontsize=8)

plt.show()
```

```python
plt.figure(figsize = (14,8))

m =np.triu(np.ones_like(wine_ds.corr(), dtype=bool))

sns.heatmap(wine_ds.corr(), mask = m, annot = True, vmin= -1.0, vmax= 1.0, center = 0, cmap = 'RdBu_r');
```
